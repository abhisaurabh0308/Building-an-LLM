```python
# üì¶ Install required packages
!pip install fitz PyMuPDF sentence-transformers qdrant-client requests pdfplumber torch torchvision --extra-index-url https://download.pytorch.org/whl/torch_stable.html --upgrade transformers -q
!pip uninstall -y fitz -q

# %% [markdown]
# **UPLOAD PDF & PARSE**
# %%
# üîº Upload PDF
from google.colab import files
import pdfplumber

uploaded = files.upload()
# Get the path of the uploaded file
pdf_path = next(iter(uploaded))

def extract_text_pdfplumber(file_path):
    """Extracts text from a PDF file using pdfplumber."""
    text = ""
    try:
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                text += page.extract_text() + "\n"
    except Exception as e:
        print(f"Error extracting text: {e}")
        return None
    return text

# Extract text from the uploaded PDF
raw_text = extract_text_pdfplumber(pdf_path)

if raw_text:
    print("‚úÖ PDF parsed. Length:", len(raw_text))
else:
    print("‚ùå Failed to parse PDF.")

# %% [markdown]
# **CHUNKING**
# %%
def chunk_text(text, max_words=200):
    """Chunks text into smaller pieces based on sentence boundaries and max word count."""
    import re
    # Split by sentence endings, keeping the delimiter attached to the sentence
    sentences = re.split(r'(?<=[.!?]) +', text)
    chunks = []
    chunk = ""

    for sentence in sentences:
        # Check if adding the next sentence exceeds the max word limit
        # Use a temporary variable to avoid modifying chunk before the check
        temp_chunk = chunk + (" " if chunk else "") + sentence
        if len(temp_chunk.split()) > max_words:
            # If adding the sentence exceeds the limit, append the current chunk
            if chunk: # Append only if chunk is not empty
                chunks.append(chunk.strip())
            chunk = sentence.strip() # Start a new chunk with the current sentence
        else:
            # Otherwise, add the sentence to the current chunk
            chunk = temp_chunk.strip()

    # Add the last chunk if it's not empty
    if chunk:
        chunks.append(chunk.strip())

    return chunks

# Ensure raw_text is defined and not None before chunking
if 'raw_text' in globals() and raw_text:
    chunks = chunk_text(raw_text)
    print(f"‚úÖ Text chunked into {len(chunks)} chunks.")
    if chunks:
        print("üìù Sample Chunk:\n", chunks[0][:500])
else:
    print("‚ùå raw_text is not defined or empty. Cannot perform chunking.")


# %% [markdown]
# **EMBEDDING**
# %%
# Ensure chunks is defined before this cell is executed
if 'chunks' in globals() and chunks:
    from sentence_transformers import SentenceTransformer

    # Load pretrained embedding model
    model = SentenceTransformer("all-MiniLM-L6-v2")

    # Generate vector embeddings for each chunk
    embeddings = model.encode(chunks, show_progress_bar=True)

    print(f"‚úÖ Generated embeddings for {len(chunks)} chunks.")
    print("üìê Embedding dimension:", len(embeddings[0]))
else:
    print("‚ùå chunks is not defined or empty. Cannot generate embeddings.")

# %% [markdown]
# **VECTOR DATABASE (QDRANT)**
# %%
# Ensure chunks and embeddings are defined before this cell is executed
if 'chunks' in globals() and chunks and 'embeddings' in globals() and embeddings is not None and len(embeddings) == len(chunks):
    from qdrant_client import QdrantClient
    from qdrant_client.models import Distance, VectorParams, PointStruct
    import uuid

    # Initialize Qdrant client
    # Using in-memory storage for simplicity
    client = QdrantClient(":memory:")

    collection_name = "my_pdf_collection"
    vector_size = len(embeddings[0]) # Get the dimension from the first embedding

    # Create collection if it doesn't exist
    try:
        client.get_collection(collection_name=collection_name)
        print(f"Collection '{collection_name}' already exists.")
    except:
        print(f"Creating collection '{collection_name}'...")
        client.recreate_collection(
            collection_name=collection_name,
            vectors_config=VectorParams(size=vector_size, distance=Distance.COSINE),
        )
        print(f"Collection '{collection_name}' created.")


    # Prepare data for upsertion
    points = []
    for i, chunk in enumerate(chunks):
        points.append(
            PointStruct(
                id=str(uuid.uuid4()), # Generate a unique ID for each point
                vector=embeddings[i].tolist(), # Convert numpy array to list
                payload={"text": chunk} # Store the original text chunk as payload
            )
        )

    # Upsert points to the collection
    try:
        operation_info = client.upsert(
            collection_name=collection_name,
            wait=True,
            points=points,
        )
        print(f"‚úÖ Upserted {len(points)} points. Status: {operation_info.status}")

        # Optional: Check collection count
        count_result = client.count(collection_name=collection_name, exact=True)
        print(f"Total points in collection: {count_result.count}")

    except Exception as e:
        print(f"‚ùå Error during upsertion: {e}")

else:
    print("‚ùå chunks or embeddings are not properly defined or matched. Cannot upsert to Qdrant.")
```

```python
# Ensure the required imports are loaded
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams, PointStruct
import uuid

# Check dimensions
print("‚úÖ Chunks:", len(chunks), "Embeddings:", len(embeddings))

# Initialize in-memory Qdrant instance
client = QdrantClient(":memory:")

COLLECTION_NAME = "pdf_chunks"

# Create the collection
client.recreate_collection(
    collection_name=COLLECTION_NAME,
    vectors_config=VectorParams(size=len(embeddings[0]), distance=Distance.COSINE)
)

# Create data points to store in vector DB
points = [
    PointStruct(
        id=str(uuid.uuid4()),
        vector=list(vector),
        payload={"text": chunk}
    )
    for chunk, vector in zip(chunks, embeddings)
]

# Upload to Qdrant
client.upsert(
    collection_name=COLLECTION_NAME,
    points=points
)

print(f"‚úÖ {len(points)} chunks stored in Qdrant.")
```

**SEARCH TOP K CHUNKS**

```python
def search_similar_chunks(query, top_k=3):
    # Convert query to embedding
    query_vector = model.encode([query])[0].tolist()

    # Search in Qdrant
    search_result = client.search(
        collection_name=COLLECTION_NAME,
        query_vector=query_vector,
        limit=top_k
    )

    # Return top matched chunks (text)
    return [hit.payload["text"] for hit in search_result]
```
